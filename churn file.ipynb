```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

# Converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("count of customer by churn")
plt.show()
```

```python
plt.figure(figsize = (3, 4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct ="%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

# from the given pie chart we can conclude that 26.54 of out customer have churned out
# now let explore the reason behind it

```python
ex = sns.countplot(x ='gender', data = df, hue = "Churn")
ex.bar_label(ex.containers[0])
plt.title("Churn by gender")
plt.show()
```

```python
ex = sns.countplot(x ='SeniorCitizen', data = df, hue = "Churn")
ex.bar_label(ex.containers[0])
plt.title("Churn by gender")
plt.show()
```

```python
categorical_columns = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]

# Subplot grid setup
n_cols = 4  # number of columns in the subplot grid
n_rows = (len(categorical_columns) + n_cols - 1) // n_cols

plt.figure(figsize=(20, n_rows * 4))

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
